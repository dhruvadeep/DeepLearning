```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
data.head()
```

```python
# Creating the X and y arrays
X = data.drop('MEDV', axis=1).values
Y = data['MEDV'].values
```

```python
# Adding a column of ones to the X array
X = np.c_[np.ones(X.shape[0]), X]
```

Calculating the value of B=(XT⋅X)−1⋅XT⋅y

```python
# XTX = np.dot(X.T, X)
# XTX_inv = np.linalg.inv(XTX)
# WB = np.dot(np.dot(XTX_inv, X.T), Y)

B = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
```

```python
# testing the model
Y_pred = X.dot(B)

# Plotting the results
plt.plot(Y[:20], label='Real')
# plt.show()
plt.plot(Y_pred[:20], label='Predicted')
plt.show()
```

```python
# Show graph per error
plt.plot(((Y - Y_pred)[:20] ** 2)**0.5, label='Error')
plt.plot([0]*20, label='Zero error')
plt.show()
```
